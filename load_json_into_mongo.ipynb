```python
import json
import os
import pymongo
import re
import string
import demjson
from bson import json_util
```

```python
dir='/Users/katiebalcewicz/capstone-spring-2018-team-7/collect_data/android'
connection = pymongo.MongoClient("mongodb://localhost")
db = connection.apps
temp = db.temp
for filename in os.listdir(dir):
    filepath = dir+'/'+filename
    if not filename == "retest_chart_2018-02-27.json":
        continue
    split = re.split('_|\.',filename)
    country = split[0]
    #print(country)
    chart = split[1]
    #print(chart)
    date = split[2]
    #print(date)
    data = open(filepath).read()
    data = data.replace("undefined","\"undefined\"")
    #print(data)
    obj = json_util.loads(data)
    #print(obj)
    #print(len(obj))
    for i in range(0, len(obj)):
        obj[i]["country"] = country
        obj[i]["chart"] = chart
        obj[i]["date"] = date
        obj[i]["rank"] = i+1
    #print(obj)
    app_id = temp.insert_many(obj)

```
